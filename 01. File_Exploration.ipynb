# Making a Dataset for Analysis

```python
#connvert binary features to numerical
binary_cols = ['potential_issue', 'deck_risk', 'oe_constraint', 'ppap_risk', 'stop_auto_buy', 
              'rev_stop', 'went_on_backorder']
df_dummies = pd.get_dummies(features, columns = binary_cols, drop_first = True)
```

```python
df_dummies.head()
```

# Train, Test Split

Making a directory "derived_data" to store our modelling data

Since we have a test dataset included, we'll just add a validation dataset to validate our results before final testing

```python
!mkdir -p derived_data
```

```python
#Split data into train and test  
train,  validate = np.split(df_dummies.sample(frac=1), [int(.8*len(df_dummies))])
```
